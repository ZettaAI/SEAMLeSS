```python
from aligner import Aligner, BoundingBox
import matplotlib.pyplot as plt
from copy import deepcopy
%matplotlib inline
s = 64*(2**7)
t = 64*(2**7)
mip = 7
source_z = 117
target_z = 116
v_off = (10240, 4096, 0)

#patch_bbox = BoundingBox(v_off[0]+t, v_off[0]+t+s, v_off[1]+t, v_off[1]+t+s, mip=0, max_mip=9)
#patch_bbox = BoundingBox(v_off[0], v_off[0]+57344, v_off[1], v_off[1]+40960, mip=0, max_mip=9)
patch_bbox = BoundingBox(v_off[0]+t, v_off[0]+t+s, v_off[1]+t, v_off[1]+t+s, mip=0, max_mip=9)
crop = 64
influence_bbox = deepcopy(patch_bbox)
influence_bbox.uncrop(crop, mip=mip)

a = Aligner('model/2_5_2.pt', (64, 64), 1024, 32, 9, 9,
      'gs://neuroglancer/pinky40_alignment/prealigned',
      'gs://neuroglancer/nflow_tests/small', move_anchor=True)
```

```python
mip = 6
src_patch = a.get_image_data(a.dst_ng_path, source_z, influence_bbox, mip)
src_prewarp = a.get_image_data(a.src_ng_path, source_z, influence_bbox, mip) 
tgt_patch = a.get_image_data(a.src_ng_path, target_z, influence_bbox, mip)
print (src_patch.shape)
```

```python
fig=plt.figure(figsize=(20, 30))

img = src_prewarp[0]
ax = plt.subplot(131)
ax.set_title("source prewarp", fontsize=26)
plt.imshow(img)

img = src_patch[0]
ax = plt.subplot(132)
ax.set_title("source postwarp", fontsize=26)
plt.imshow(img)

img = tgt_patch[0]
ax = plt.subplot(133)
ax.set_title("target", fontsize=26)
plt.imshow(img)

plt.show()
```

```python
mip=6
chunks = a.break_into_chunks(patch_bbox, a.dst_chunk_sizes[mip], a.dst_voxel_offsets[mip], mip)
c0 = chunks[0]
ic0 = deepcopy(c0)
ic0.uncrop(a.max_displacement, mip=0)

mip=9
res = a.get_abs_residual(source_z, ic0, mip=mip)
print (res.shape)
```

```python
fig=plt.figure(figsize=(10, 20))
ax = plt.subplot(121)
ax.set_title("x residual", fontsize=26)
plt.imshow(res[0, :, :, 0])
print(res[0,0,0,0])


ax = plt.subplot(122)
ax.set_title("y residual", fontsize=26)
plt.imshow(res[0, :, :, 1])
print(res[0,0,0,1])
```

```python
from util import warp, upsample
import numpy as np

patch = patch_bbox

mip = 9
src_data = a.get_image_data(a.src_ng_path, source_z, patch, mip)
tgt_data = a.get_image_data(a.dst_ng_path, target_z, patch, mip)
rel_flow = a.net.process(src_data, tgt_data, mip)

mip = 6
scale_factor = 2**(9-6)
rel_flow_up = np.stack((upsample(rel_flow[:, :, :, 0], scale_factor),
                        upsample(rel_flow[:, :, :, 1], scale_factor)), axis=3)


agg_flow = patch.identity(mip=mip) 
agg_flow = np.expand_dims(agg_flow, axis=0)
agg_flow += rel_flow_up

raw_data = a.get_image_data(a.src_ng_path, source_z, patch, mip)
raw_tgt = a.get_image_data(a.src_ng_path, target_z, patch, mip)
warped = warp(raw_data, agg_flow)

```

```python
print (np.max(rel_flow_up))
fig=plt.figure(figsize=(10, 20))
ax = plt.subplot(121)
ax.set_title("x residual", fontsize=26)
plt.imshow(rel_flow_up[0, :, :, 0])

fig=plt.figure(figsize=(10, 20))
ax = plt.subplot(122)
ax.set_title("x flow", fontsize=26)
plt.imshow(agg_flow[0, :, :, 0])
plt.show()
```
