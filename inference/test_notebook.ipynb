```python
from aligner import Aligner, BoundingBox
import matplotlib.pyplot as plt
from copy import deepcopy
%matplotlib inline

s = 256
t = 256
mip = 6
source_z = 11
target_z = 10
v_off = (160, 64, 0)

patch_bbox = BoundingBox(v_off[0]+t, v_off[0]+t+s, v_off[1]+t, v_off[1]+t+s, mip=mip)
crop = 30
influence_bbox = deepcopy(patch_bbox)
influence_bbox.uncrop(crop, mip=mip)

a = Aligner('model/1_6_3.pt', (s, s), s, 6, 4, 'gs://neuroglancer/pinky40_v11/image', 'gs://neuroglancer/nflow_tests/p40') 
```

```python
fig=plt.figure(figsize=(10, 20))

columns = 2
rows = 1

img = a.get_aggregate_flow(target_z, patch_bbox, mip)[0, :, :, 0]
fig.add_subplot(1, 2, 1)
plt.imshow(img)

img = a.get_aggregate_flow(target_z, patch_bbox, mip)[0, :, :, 1]
fig.add_subplot(1, 2, 2)
plt.imshow(img)

plt.show()
```

```python
src_patch = a.get_warped_patch(a.src_ng_path, source_z, influence_bbox, mip)
src_prewarp = a.get_image_data(a.src_ng_path, source_z, influence_bbox, mip) 
tgt_patch = a.get_image_data(a.src_ng_path, target_z, influence_bbox, mip)
print (src_patch.shape)
```

```python
fig=plt.figure(figsize=(20, 30))

img = src_prewarp[0]
ax = plt.subplot(131)
ax.set_title("source prewarp", fontsize=26)
plt.imshow(img)

img = src_patch[0]
ax = plt.subplot(132)
ax.set_title("source postwarp", fontsize=26)
plt.imshow(img)

img = tgt_patch[0]
ax = plt.subplot(133)
ax.set_title("target", fontsize=26)
plt.imshow(img)

plt.show()
```

```python
fig=plt.figure(figsize=(10, 20))

ax = plt.subplot(121)
ax.set_title("x residual", fontsize=26)
plt.imshow(x_res)

img = res[0,:,:,1]
ax = plt.subplot(122)
ax.set_title("y residual", fontsize=26)
plt.imshow(y_res)
```
