```python
from aligner import Aligner, BoundingBox
import matplotlib.pyplot as plt
from copy import deepcopy
%matplotlib inline

s = 512
t = 5120
mip = 2
source_z = 11
target_z = 10
v_off = (2560, 1024, 0)

patch_bbox = BoundingBox(v_off[0]+t, v_off[0]+t+s, v_off[1]+t, v_off[1]+t+s, mip=mip, max_mip=6)
crop = 64
influence_bbox = deepcopy(patch_bbox)
influence_bbox.uncrop(crop, mip=mip)

a = Aligner('model/1_6_3.pt', (256, 256), 512, 64, 6, 4, 'gs://neuroglancer/pinky40_v11/image', 'gs://neuroglancer/nflow_tests/p40') 
```

```python
src_patch = a.get_warped_patch(a.src_ng_path, source_z, influence_bbox, mip)
src_prewarp = a.get_image_data(a.src_ng_path, source_z, influence_bbox, mip) 
tgt_patch = a.get_image_data(a.src_ng_path, target_z, influence_bbox, mip)
print (src_patch.shape)
```

```python
fig=plt.figure(figsize=(20, 30))

img = src_prewarp[0]
ax = plt.subplot(131)
ax.set_title("source prewarp", fontsize=26)
plt.imshow(img)

img = src_patch[0]
ax = plt.subplot(132)
ax.set_title("source postwarp", fontsize=26)
plt.imshow(img)

img = tgt_patch[0]
ax = plt.subplot(133)
ax.set_title("target", fontsize=26)
plt.imshow(img)

plt.show()
```

```python
a.compute_section_pair_residuals(source_z, target_z, patch_bbox)
```

```python
a.render_section(source_z, patch_bbox, mip=2)
```

```python
x_res = a.get_x_residual(source_z, patch_bbox, mip=6)
y_res = a.get_y_residual(source_z, patch_bbox, mip=6)
print (x_res.shape)
fig=plt.figure(figsize=(10, 20))
ax = plt.subplot(121)
ax.set_title("x residual", fontsize=26)
plt.imshow(x_res[0])


ax = plt.subplot(122)
ax.set_title("y residual", fontsize=26)
plt.imshow(y_res[0])
```

```python
agg_flow = a.get_aggregate_flow(source_z, patch_bbox, mip=2)
x_res = agg_flow[..., 0]
y_res = agg_flow[..., 1]
print (x_res.shape)
fig=plt.figure(figsize=(10, 20))
ax = plt.subplot(121)
ax.set_title("x residual", fontsize=26)
plt.imshow(x_res[0])


ax = plt.subplot(122)
ax.set_title("y residual", fontsize=26)
plt.imshow(y_res[0])
```
