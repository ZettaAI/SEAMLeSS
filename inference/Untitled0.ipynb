```python
from aligner import Aligner, BoundingBox
import matplotlib.pyplot as plt
from copy import deepcopy
import numpy as np
%matplotlib inline

s = 5120
t = 5120
mip = 2
source_z = 11
target_z = 10
v_off = (2560, 1024, 0)
patch_bbox = BoundingBox(v_off[0]+t, v_off[0]+t+s, v_off[1]+t, v_off[1]+t+s, mip=mip)
crop = 30
influence_bbox = deepcopy(patch_bbox)
influence_bbox.uncrop(crop, mip=mip)

a = Aligner('model/1_6_3.pt', (1024, 1024), 512, 16, 6, 4, 'gs://neuroglancer/pinky40_v11/image', 'gs://neuroglancer/nflow_tests/p40') 
```

```python
src_patch = a.get_warped_patch(a.src_ng_path, source_z, influence_bbox, 5)
src_prewarp = a.get_image_data(a.src_ng_path, source_z, influence_bbox, 5) 
tgt_patch = a.get_image_data(a.src_ng_path, target_z, influence_bbox, 5)
print (src_patch.shape)
```

```python
fig=plt.figure(figsize=(20, 30))

img = src_prewarp[0]
ax = plt.subplot(131)
ax.set_title("source prewarp", fontsize=26)
plt.imshow(img)

img = src_patch[0]
ax = plt.subplot(132)
ax.set_title("source postwarp", fontsize=26)
plt.imshow(img)

img = tgt_patch[0]
ax = plt.subplot(133)
ax.set_title("target", fontsize=26)
plt.imshow(img)

plt.show()
```
