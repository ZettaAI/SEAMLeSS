## Create UNet ##

```python
# helper operations 
def conv3x3(in_channels, out_channels):
    return nn.Conv2d(in_channels, out_channels,
        kernel_size=3, stride=1, padding=1, bias=True)

def maxpool2x2():
    return nn.MaxPool2d(kernel_size=2, stride=2, padding=0)

class UpConv2x2(nn.Module):
    def __init__(self, channels):
        super(UpConv2x2, self).__init__()
        self.upsample = nn.Upsample(scale_factor=2)
        self.conv = nn.Conv2d(channels, channels // 2,
            kernel_size=2, stride=1, padding=0, bias=True)

    def forward(self, x):
        x = self.upsample(x)
        x = F.pad(x, (0,1,0,1))
        x = self.conv(x)
        return x 
        
def concat(xh, xv):
    return torch.cat([xh, xv], dim=1)


# unet blocks
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        """
        Args:
            in_channels: number of channels in input (1st) feature map
            out_channels: number of channels in output feature maps
        """
        super(ConvBlock, self).__init__()
        self.conv1 = conv3x3(in_channels, out_channels)
        self.conv2 = conv3x3(out_channels, out_channels)
        self.conv3 = conv3x3(out_channels, out_channels)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        return x

class DownConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        """
        Args:
            in_channels: number of channels in input (1st) feature map
            out_channels: number of channels in output feature maps
        """
        super(DownConvBlock, self).__init__()
        self.maxpool = maxpool2x2()
        self.conv1 = conv3x3(in_channels, out_channels)
        self.conv2 = conv3x3(out_channels, out_channels)
        self.conv3 = conv3x3(out_channels, out_channels)

    def forward(self, x):
        x = self.maxpool(x)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        return x

class UpConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        """
        Args:
            in_channels: number of channels in input (1st) feature map
            out_channels: number of channels in output feature maps
        """
        super(UpConvBlock, self).__init__()
        self.upconv = UpConv2x2(in_channels)
        self.conv1 = conv3x3(in_channels, out_channels)
        self.conv2 = conv3x3(out_channels, out_channels)
        self.conv3 = conv3x3(out_channels, out_channels)

    def forward(self, xh, xv):
        """
        Args:
            xh: torch Variable, activations from same resolution feature maps (gray arrow in diagram)
            xv: torch Variable, activations from lower resolution feature maps (green arrow in diagram)
        """
        xv = self.upconv(xv)
        x = concat(xh, xv)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        return x

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        fs = [16,32,64,128,256]
        self.conv_in = ConvBlock(1, fs[0])
        self.dconv1 = DownConvBlock(fs[0], fs[1])
        self.dconv2 = DownConvBlock(fs[1], fs[2])
        self.dconv3 = DownConvBlock(fs[2], fs[3])
        self.dconv4 = DownConvBlock(fs[3], fs[4])

        self.uconv1 = UpConvBlock(fs[4], fs[3])
        self.uconv2 = UpConvBlock(fs[3], fs[2])
        self.uconv3 = UpConvBlock(fs[2], fs[1])
        self.uconv4 = UpConvBlock(fs[1], fs[0])
        self.conv_out = conv3x3(fs[0], 1)

        self._initialize_weights()

    def forward(self, x):
        x1 = self.conv_in(x)
        x2 = self.dconv1(x1)
        x3 = self.dconv2(x2)
        x4 = self.dconv3(x3)
        x5 = self.dconv4(x4)
        x6 = self.uconv1(x4, x5)
        x7 = self.uconv2(x3, x6)
        x8 = self.uconv3(x2, x7)
        x9 = self.uconv4(x1, x8)
        x10 = self.conv_out(x9)
        x11 = self.conv_out(x9)
        return x10, x11

    def _initialize_weights(self):
        conv_modules = [m for m in self.modules() if isinstance(m, nn.Conv2d)]
        for m in conv_modules:
          n = m.kernel_size[0] * m.kernel_size[1] * m.out_channels
          m.weight.data.normal_(0, np.sqrt(2. / n))
```

## Train UNet ##

```python
# train utils
def np2var(arr, cuda):
    arr = np.expand_dims(np.expand_dims(arr, 0), 0)
    arr = torch.from_numpy(arr)
    arr = Variable(arr)
    if cuda:
        arr = arr.cuda()
    return arr

def var2np(arr):
    arr = arr.data.cpu().numpy()[0,0]
    return arr
    
```

```python
# plot utils
def plot_learning(plot_handles, ylabel):
    plt.legend(handles=plot_handles, loc="center right")
    plt.xlabel("Iterations")
    plt.ylabel(ylabel)

def plot(step, log_every, train_loss, val_loss, img_trn, img_val, crack_trn, fold_trn, 
                                                            pred_crack_val, true_crack_trn, true_crack_val,
                                                            pred_fold_val, true_fold_trn, true_fold_val):
    clear_output(wait=False)
    plt.gcf().clear()
    
    ax = plt.subplot(441)
    iters = np.arange(0,len(train_loss))*log_every
    train_plot, = plt.plot(iters, train_loss, 'r', label="training")
    val_plot, = plt.plot(iters, val_loss, 'b', label="validation")
    plot_learning([train_plot, val_plot], "Loss")
    plt.title("Cross entropy loss")
    ax.set_yscale("log")

    plt.subplot(442)
    plt.imshow(img_trn, cmap='gray')
    plt.title("Input image, train")
    
    plt.subplot(443)
    plt.imshow(crack_trn, cmap='gray')
    plt.title("Predicted crack, train")
    
    plt.subplot(444)
    plt.imshow(fold_trn, cmap='gray')
    plt.title("Predicted fold, train")
    
    
    plt.subplot(447)
    plt.imshow(true_crack_trn, cmap='gray')
    plt.title("True crack, train")
    
    plt.subplot(448)
    plt.imshow(true_fold_trn, cmap='gray')
    plt.title("True fold, train")
    
    
    
    
    plt.subplot(4,4,10)
    plt.imshow(img_val, cmap='gray')
    plt.title("Input image, val")
    
    plt.subplot(4,4,11)
    plt.imshow(pred_crack_val, cmap='gray')
    plt.title("Predicted crack, val")
    
    plt.subplot(4,4,12)
    plt.imshow(pred_fold_val, cmap='gray')
    plt.title("Predicted fold, val")
    
    plt.subplot(4,4,15)
    plt.imshow(true_crack_val, cmap='gray')
    plt.title("True crack, val")
    
    plt.subplot(4,4,16)
    plt.imshow(true_fold_val, cmap='gray')
    plt.title("True fold, val")


    plt.gcf().set_size_inches(18, 7)
    display(plt.gcf())
```

```python
## Train network ##

# params
USECUDA = True
max_iter = 200
log_every = 50
plot_every = 200
lr = 0.00005

# create net
net = UNet()
if USECUDA:
    net.cuda()
    
# create optimizer 
optimizer = optim.Adam(net.parameters(), lr=lr)

# train net
losses_trn = []
losses_val = []
for t in range(max_iter):
    # choose sample
    idx = np.random.randint(0,images_trn.shape[0])
    while cracks_trn_exclude[idx] & folds_trn_exclude[idx]:
        idx = np.random.randint(0,images_trn.shape[0])
    img, crack, fold = images_trn[idx], cracks_trn[idx], folds_trn[idx]
    crack_train_include = np2var(np.array((~cracks_trn_exclude[idx])).astype(np.float32), cuda=USECUDA)
    fold_train_include = np2var(np.array((~folds_trn_exclude[idx])).astype(np.float32), cuda=USECUDA)
    
    # flip
    p = np.random.randint(0,3)
    if p > 1:
        img = np.flip(img,axis=1).copy()
        crack = np.flip(crack,axis=1).copy()
        fold = np.flip(fold,axis=1).copy()
    if p > 0:
        img = np.flip(img,axis=0).copy()
        crack = np.flip(crack,axis=0).copy()
        fold = np.flip(fold,axis=0).copy()
    
    # format
    img = np2var(img, cuda=USECUDA)
    crack = np2var(crack, cuda=USECUDA)
    fold = np2var(fold, cuda=USECUDA)
    
    # forward, backward, update
    optimizer.zero_grad()
    pred_crack, pred_fold = net(img)
    loss_crack = F.binary_cross_entropy_with_logits(pred_crack, crack)
    loss_fold = F.binary_cross_entropy_with_logits(pred_fold, fold)
    loss = loss_crack*crack_train_include + loss_fold*fold_train_include
    loss.backward()
    optimizer.step()

    # log stuff
    if t % log_every == 0: 
        # validation info
        idx = np.random.randint(0,images_val.shape[0])
        while cracks_val_exclude[idx] & folds_val_exclude[idx]:
            idx = np.random.randint(0,images_val.shape[0])
        img_val, crack_val, fold_val = images_val[idx], cracks_val[idx], folds_val[idx]
        crack_val_include = np2var(np.array((~cracks_val_exclude[idx])).astype(np.float32), cuda=USECUDA)
        fold_val_include = np2var(np.array((~folds_val_exclude[idx])).astype(np.float32), cuda=USECUDA)

        # flip
        p = np.random.randint(0,3)
        if p > 1:
            img = np.flip(img,axis=1).copy()
            crack = np.flip(crack,axis=1).copy()
            fold = np.flip(fold,axis=1).copy()
        if p > 0:
            img = np.flip(img,axis=0).copy()
            crack = np.flip(crack,axis=0).copy()
            fold = np.flip(fold,axis=0).copy()
        
        # format
        img_val = np2var(img_val, cuda=USECUDA)
        crack_val = np2var(crack_val, cuda=USECUDA)
        fold_val = np2var(fold_val, cuda=USECUDA)

        # forward
        pred_crack_val, pred_fold_val = net(img_val)
        loss_crack_val = F.binary_cross_entropy_with_logits(pred_crack_val, crack_val)
        loss_fold_val = F.binary_cross_entropy_with_logits(pred_fold_val, fold_val)
        loss_val = loss_crack_val*crack_val_include + loss_fold_val*fold_val_include

        # store
        losses_trn.append(loss.data[0])
        losses_val.append(loss_val.data[0])
        
    # plot stuff
    if t % plot_every == 0:
        plot(t, log_every, losses_trn, losses_val, 
             var2np(img), var2np(img_val),
             var2np(F.sigmoid(pred)), var2np(F.sigmoid(pred_val)),
             var2np(label), var2np(label_val))
    
    
```
